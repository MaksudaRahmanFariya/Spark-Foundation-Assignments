```python
data = pd.read_csv("../input/breastcancer-dataset/data.csv")
data.info()
```

```python
data.head()
```

```python
data1= data.copy()
data=data.drop(['id','Unnamed: 32'],axis=1)
```

```python
y = data['diagnosis']
x = data.drop('diagnosis',axis=1)
```

```python
x.shape , y.shape
```

```python
log_reg = LogisticRegression()
```

```python
help(LogisticRegression)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.02,random_state=11)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
log_reg = LogisticRegression(solver='saga',penalty ='l1')
log_reg.fit(xtrain,ytrain)
```

```python
xtrain
```

```python
pred = log_reg.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
report = classification_report(ytest,pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(ytest,pred)
print(cm)
```

```python
print("Accuracy score:",accuracy_score(ytest,pred))
```
